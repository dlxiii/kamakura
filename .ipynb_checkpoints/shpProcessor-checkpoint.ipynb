```python
def houseSumInArea(rec_h0, id_area, flag):
    houseSum = 0
    for i in rec_h0:
        if i[flag+1] == id_area:
            houseSum = houseSum + i[1] 
    return round(houseSum,2)
```

```python
def houseListInArea(rec_h0, id_area, flag):
    houseList = []
    for i in rec_h0:
        if i[flag+1] == id_area:
            houseList.append(i[0])
    return houseList
```

```python
def houseNumInArea(rec_h0, id_area, flag):
    houseNum = 0
    for i in rec_h0:
        if i[flag+1] == id_area:
            houseNum = houseNum + 1
    return houseNum
```

```python
def houseInfoStatic(rec_h0, rec_xx):
    idList = [i[0] for i in rec_xx]
    if len(idList) == 46:
        flag = 1
    if len(idList) == 219:
        flag = 2
    outputData = []
    for j in idList:
        houseSum = houseSumInArea(rec_h0, j, flag)
        houseList = houseListInArea(rec_h0, j, flag)
        houseNum = houseNumInArea(rec_h0, j, flag)
        outputData.append([j,houseNum,houseSum,houseList])
    return outputData
```

```python
for i in [rec_d1,rec_d2]:
    if len(i) == 46: flag = 1
    if len(i) == 219: flag = 2
    df = pd.DataFrame({'Inlet':[], 'House number':[], 'House roof area':[], 'House ID':[]})
    list_0 = []
    list_1 = []
    list_2 = []
    list_3 = []
    for j in houseInfoStatic(rec_h0, i):
        list_0.append(j[0])
        list_1.append(j[1])
        list_2.append(j[2])
        list_3.append(j[3])
    df['Inlet'] = list_0
    df['House number'] = list_1
    df['House roof area'] = list_2
    df['House ID'] = list_3
    pickle.dump(df, open('otps/houseInfoStatic_'+str(flag)+'.data', mode = 'wb'), protocol = True)
```

```python
def isRayIntersectsSegment(poi,s_poi,e_poi): #[x,y] [lng,lat]
    # https://www.jianshu.com/p/ba03c600a557
    # [lng,lat]
    if s_poi[1]==e_poi[1]: #
        return False
    if s_poi[1]>poi[1] and e_poi[1]>poi[1]: #
        return False
    if s_poi[1]<poi[1] and e_poi[1]<poi[1]: #
        return False
    if s_poi[1]==poi[1] and e_poi[1]>poi[1]: #spoint
        return False
    if e_poi[1]==poi[1] and s_poi[1]>poi[1]: #epoint
        return False
    if s_poi[0]<poi[0] and e_poi[1]<poi[1]: #
        return False
    xseg=e_poi[0]-(e_poi[0]-s_poi[0])*(e_poi[1]-poi[1])/(e_poi[1]-s_poi[1]) #
    if xseg<poi[0]: #
        return False
    return True  #
def isPoiWithinPoly(poi,poly):
    # https://www.jianshu.com/p/ba03c600a557
    #
    #poly=[[[x1,y1],[x2,y2],,[xn,yn],[x1,y1]],[[w1,t1],[wk,tk]]] 

    # 
    #if not isPoiWithinBox(poi,mbr=[[0,0],[180,90]]): return False
    #
    sinsc = 0 #
    epoly = poly
    #for epoly in poly: #->each polygon [[x1,y1],[xn,yn]]
    for i in range(len(epoly)-1): #[0,len-1]
        s_poi = epoly[i]
        e_poi = epoly[i+1]
        if isRayIntersectsSegment(poi,s_poi,e_poi):
             sinsc += 1 #1
    return True if sinsc%2 == 1 else  False
```
