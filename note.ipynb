```python
import shapefile as sf
import matplotlib.pyplot as plt
```

```python
file1 = '1_level_drainage.shp'
file2 = '2_level_drainage.shp'
```

```python
spf_1 = sf.Reader(file1)
for sh in spf_1.shapeRecords():
    x = [i[0] for i in sh.shape.points[:]]
    y = [i[1] for i in sh.shape.points[:]]
    plt.plot(x,y)
plt.show()
```

```python
spf_2 = sf.Reader(file2)
for sh in spf_2.shapeRecords():
    x = [i[0] for i in sh.shape.points[:]]
    y = [i[1] for i in sh.shape.points[:]]
    plt.plot(x,y)
plt.show()
```
